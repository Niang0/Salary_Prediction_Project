Pourquoi créer main() ?

Tu pourrais écrire directement :

df = load_data(...)
...

sans fonction.

Mais en Python, la bonne pratique est :

def main():
    ...

Pourquoi ?

Parce que plus tard :

Docker exécutera main()
Streamlit pourra appeler main()
FastAPI pourra appeler main()
les tests pourront appeler main()

Une seule entrée pour lancer le projet.

La ligne la plus célèbre de Python
if __name__ == "__main__":
    main()

Au début, elle paraît mystérieuse.

En réalité, elle signifie :

"Si ce fichier est exécuté directement, lance la fonction main()."

Par exemple :

python main.py

➡️ main() est exécutée.

En revanche, si un autre fichier fait :

import main

alors main() ne se lance pas automatiquement.

C'est un mécanisme qui permet de réutiliser du code sans provoquer d'exécution non désirée.

Une analogie

Imagine un restaurant.

Le chef prépare les plats.

Les serveurs servent les plats.

Le caissier encaisse.

Le directeur ne cuisine pas.

Il coordonne tout.

Le main.py est le directeur.

Les autres fichiers réalisent le travail.
